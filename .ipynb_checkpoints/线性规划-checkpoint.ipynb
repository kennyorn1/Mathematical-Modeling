```python
c=np.array([2,3,-5])
A=np.array([[-2,5,-1],[1,3,1]])
b=np.array([-10,12])
Aeq=np.array([[1,1,1]])
beq=np.array([7])
```

```python
res=optimize.linprog(-c,A,b,Aeq,beq)
print(res)
```

```python
print(res.fun)
```

```python
print(res.x)
```
